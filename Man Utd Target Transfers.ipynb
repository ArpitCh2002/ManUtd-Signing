```python
import pandas as pd

# Load the Excel file
file_path = "D:\Hackathon Project\PL_Player_Dtaa.xlsx"  # replace with your actual file name
xls = pd.ExcelFile(file_path)

print(xls.sheet_names)
```

```python
df = pd.read_excel(xls, sheet_name='Premier League - All Stats')

# Quick check
df.head()
```

```python
# Define position-based filters
midfield_positions = ['CM', 'DM', 'AM', 'CAM', 'CDM', 'LCM', 'RCM', 'MF']
defender_positions = ['CB', 'RB', 'LB', 'LCB', 'RCB', 'RWB', 'LWB', 'DF']

# Filter midfielders and defenders
midfielders_df = df[
    df['Pos'].isin(midfield_positions) &
    (df['Squad'] != 'Manchester Utd')
].copy()

defenders_df = df[
    df['Pos'].isin(defender_positions) &
    (df['Squad'] != 'Manchester Utd')
].copy()

# Check output
print("Midfielders:", midfielders_df.shape)
print("Defenders:", defenders_df.shape)
```

PrgP, Cmp%, TotDist, PrgDist

Ball Recovery & Duel Success:

TklW, Tkl%, Tkl Def 3rd, Touches Def 3rd

Distribution Under Pressure:

Cmp, Att, Touches, Carries, CPA

```python
midfielder_columns = [
    'Player', 'Squad', 'Pos', 'PrgC', 'PrgP', 'PrgR', 'xAG', 'KP',
    'Cmp%', 'Tkl', 'Tkl+Int', 'Blocks', 'Int', 
    'Carries', 'PrgDist', 'Touches', 'Dis', 'Mis'
]

# Define defender attributes
defender_columns = [
    'Player', 'Squad', 'Pos', 'Tkl', 'Clr', 'Blocks', 'Int', 'Tkl+Int', 'Err',
    'PrgP', 'Cmp%', 'TotDist', 'PrgDist', 'TklW', 'Tkl%', 'Touches', 'Carries', 'CPA'
]

# Filter new analysis-ready DataFrames
midfielders_filtered = midfielders_df[midfielder_columns].copy()
defenders_filtered = defenders_df[defender_columns].copy()

# View data
midfielders_filtered.head()
```

```python
# Fill NaN with 0 or use mean imputation if needed
midfielders_filtered.fillna(0, inplace=True)
defenders_filtered.fillna(0, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Drop identifiers temporarily
mid_id_cols = midfielders_filtered[['Player', 'Squad', 'Pos']]
def_id_cols = defenders_filtered[['Player', 'Squad', 'Pos']]

# Normalize performance metrics
scaler = MinMaxScaler()

mid_scaled = pd.DataFrame(
    scaler.fit_transform(midfielders_filtered.drop(['Player', 'Squad', 'Pos'], axis=1)),
    columns=midfielders_filtered.drop(['Player', 'Squad', 'Pos'], axis=1).columns
)
mid_scaled = pd.concat([mid_id_cols.reset_index(drop=True), mid_scaled], axis=1)

def_scaled = pd.DataFrame(
    scaler.fit_transform(defenders_filtered.drop(['Player', 'Squad', 'Pos'], axis=1)),
    columns=defenders_filtered.drop(['Player', 'Squad', 'Pos'], axis=1).columns
)
def_scaled = pd.concat([def_id_cols.reset_index(drop=True), def_scaled], axis=1)
```

```python
mid_scaled
```

```python
# You can give equal weights or use custom weights
mid_scaled['Score'] = mid_scaled.drop(['Player', 'Squad', 'Pos'], axis=1).mean(axis=1)
def_scaled['Score'] = def_scaled.drop(['Player', 'Squad', 'Pos'], axis=1).mean(axis=1)

# Sort top performers
top_mids = mid_scaled.sort_values(by='Score', ascending=False).head(10)
top_defs = def_scaled.sort_values(by='Score', ascending=False).head(10)
```

```python
top_mids
```

```python
# Average score of all midfielders (excluding identifiers)
league_avg_mid_score = mid_scaled['Score'].mean()

# Add it as a column in top_mids for comparison
top_mids['League Avg Score'] = league_avg_mid_score

# Create a difference column
top_mids['Above Avg Diff'] = top_mids['Score'] - top_mids['League Avg Score']

# Optional: Format it for better readability
top_mids['Above Avg Diff'] = top_mids['Above Avg Diff'].round(3)

# View results
print(top_mids[['Player', 'Score', 'League Avg Score', 'Above Avg Diff']])
```

```python
# Create a compact comparison DataFrame for top midfielders
top_mids_compared = top_mids[['Player', 'Squad', 'Score', 'League Avg Score', 'Above Avg Diff']].copy()

# View the result
top_mids_compared
```

```python
# Average score of all defenders
league_avg_def_score = def_scaled['Score'].mean()

# Add it to top defenders
top_defs['League Avg Score'] = league_avg_def_score
top_defs['Above Avg Diff'] = top_defs['Score'] - top_defs['League Avg Score']
top_defs['Above Avg Diff'] = top_defs['Above Avg Diff'].round(3)

# View
print(top_defs[['Player', 'Score', 'League Avg Score', 'Above Avg Diff']])
```

```python
# Create a compact comparison DataFrame for top defenders
top_defs_compared = top_defs[['Player', 'Squad', 'Score', 'League Avg Score', 'Above Avg Diff']].copy()

# View the result
top_defs_compared
```
