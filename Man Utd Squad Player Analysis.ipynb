## Creating Impact Score Formula

### Consider three types of metrics:

🟢 Positive Impact	Goals, Assists, xG, xAG, MotM, Rating	Direct contributions to team success

🟡 Involvement	Mins, Apps (Sub), Starts, PrgP, AerialsWon	Shows how active the player is

🔴 Negative Impact	Mistakes, Red, Yel, CrdY, CrdR	Penalizes errors and discipline issues

### Reflect Real-World Importance
Some stats naturally have more impact on the game than others.

Stat	Importance	Weight Rationale

Goals	Very High	Worth 4 points — goals are the most valuable action

Assists	High	Worth 3 points — a direct contribution

xG / xAG	Medium	2 points — shows chances created or taken, even if not converted

MotM	Medium	2 points — judged impactful by external experts

Rating	Medium	2 points — composite stat, already scaled

SpG, PrgP	Lower	1.2–1.5 — shows activity and involvement, but less decisive

Mistakes/Red Cards	Negative	Higher weights penalize bad plays (Red card = -3)

### Balance Different Scales

Different columns have different natural ranges:

Goals = 0–15

AerialsWon = ~0–3 per game

Rating = ~6.0–8.5

```python
import pandas as pd

# OPTIONAL: Make sure dataframe_final is a deep copy of the original
dataframe_final = dataframe_final.copy()

# Step 1: List of columns used in the formula
impact_columns = [
    'Goals', 'Assists', 'xG', 'xAG', 'MotM', 'Rating',
    'AerialsWon', 'SpG', 'PrgP', 'PrgC',
    'Mistakes', 'Red', 'Yel', 'CrdY', 'CrdR', '90s'
]

# Step 2: Remove duplicates from columns (fixes merge-related issues)
dataframe_final = dataframe_final.loc[:, ~dataframe_final.columns.duplicated()].copy()

# Step 3: Convert relevant columns to numeric (coerce invalid to NaN)
for col in impact_columns:
    if col in dataframe_final.columns:
        dataframe_final.loc[:, col] = pd.to_numeric(dataframe_final[col], errors='coerce')

# Step 4: Replace all NaN with 0
dataframe_final.fillna(0, inplace=True)

# Step 5: Calculate the Impact Score
dataframe_final['Impact_Score'] = (
    (dataframe_final['Goals'] * 4) +
    (dataframe_final['Assists'] * 3) +
    (dataframe_final['xG'] * 2) +
    (dataframe_final['xAG'] * 2) +
    (dataframe_final['MotM'] * 2) +
    (dataframe_final['Rating'] * 2) +
    (dataframe_final['AerialsWon'] * 1.5) +
    (dataframe_final['SpG'] * 1.5) +
    (dataframe_final['PrgP'] * 1.2) +
    (dataframe_final['PrgC'] * 1.2)
) / (dataframe_final['90s'] + 1) - (
    (dataframe_final['Mistakes'] * 2) +
    (dataframe_final['Red'] * 3) +
    (dataframe_final['Yel'] * 1.5) +
    (dataframe_final['CrdY'] * 1.5) +
    (dataframe_final['CrdR'] * 3)
)

# Step 6: Final Output
df_impact_score = dataframe_final[['Player', 'Impact_Score']].copy()

df_impact_score.sort_values(by='Impact_Score', ascending=False, inplace=True)
df_impact_score
```

```python
# Define a custom styling function
def highlight_impact_score(val):
    if val <= -5:
        return 'background-color: red; color: white'
    elif -5 < val <= 5:
        return 'background-color: yellow; color: black'
    elif val > 5:
        return 'background-color: green; color: white'
    else:
        return ''

# Apply styling only to the Impact_Score column
styled_df = df_impact_score.style.applymap(highlight_impact_score, subset=['Impact_Score'])

# To display in Jupyter Notebook
styled_df
```

```python
dataframe_injuryage = pd.merge(df_injuryage, df_injuries, on='Player', how='outer')
dataframe_injuryage.fillna(0, inplace=True)

# Function to extract the whole number from 'Age_x' (e.g., '28-352' → 28)
def extract_age(age):
    if isinstance(age, str) and '-' in age:
        return int(age.split('-')[0])
    elif isinstance(age, (int, float)):
        return int(age)
    else:
        return None

# Apply the function to the Age_x column
df_injuryage['Age'] = df_injuryage['Age'].apply(extract_age)

dataframe_injuryage.drop(columns=['Age_y'], inplace=True)


dataframe_injuryage
```

### Criteria for Classification

🔹 Age Evaluation:

Age Range	age_value	age_consideration

15–24	"High Potential"	0

25–32	"Good to Have"	1

33+	"Consider Moving"	2

🔹 Injury Evaluation:

Missed Matches	injury_impact	injury_consideration

0–8	"No Concerns"	0

9–17	"Concerning"	1

18+	"Immediate Action"	2

```python
# Make a copy of the injury-age dataframe
df_impacteval = dataframe_injuryage.copy()

# AGE classification
def classify_age(age):
    if 15 <= age <= 24:
        return "High Potential", 0
    elif 25 <= age <= 32:
        return "Good to Have", 1
    elif age >= 33:
        return "Consider Moving", 2
    else:
        return "Unknown", None

df_impacteval[['age_value', 'age_consideration']] = df_impacteval['Age_x'].apply(
    lambda age: pd.Series(classify_age(age))
)

# INJURY classification
def classify_injury(missed):
    if 0 <= missed <= 8:
        return "No Concerns", 0
    elif 9 <= missed <= 17:
        return "Concerning", 1
    elif missed >= 18:
        return "Immediate Action", 2
    else:
        return "Unknown", None

df_impacteval[['injury_impact', 'injury_consideration']] = df_impacteval['Missed matches'].apply(
    lambda missed: pd.Series(classify_injury(missed))
)

# View result
#print(df_impacteval[['Player', 'Age_x', 'Missed matches', 'age_value', 'age_consideration', 'injury_impact', 'injury_consideration']].head())
df_impacteval
```
